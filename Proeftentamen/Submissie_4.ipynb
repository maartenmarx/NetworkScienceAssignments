# Verzamelingen   (2pt)

1. Definieer het product van 2 verzamelingen in Python.
2. Definieer het symmetrishe verschil van 2 verzamelingen in Python. Gebruik desnoods alleen de set functie union/vereniging. Doe het echt zelf.
3. Definieer de verzamelingen getallen groter dan 0 en kleiner dan $n$ die deelbaar zijn door $a$ maar niet door $b$. 
4. DSefinieer de functie `jaccard(A,B)` voor 2 verzamelingen A en B, met het volgende resultaat (je weet wel dat $|A|$ d cardinaliteit van $A$ geeft (wat een moeilijk woord is voor het aantal elementen dat erin zit).
$$ jaccard(A,B)= \frac{|A\cap B|}{|A\cup B|}$$

#### Maak zelf goede testjes

```python
from typing import Any


def product(A: set[Any], B: set[Any]) -> set[Any]:
    """
    Definieer het product van 2 verzamelingen in Python.
    """
    product: set[Any] = set()

    for element in A:
        for element_ in B:
            combinatie: tuple[Any, Any] = (element, element_)
            product.add(combinatie)
            
    return product


def symmetriesch_verschil(A: set[Any], B: set[Any]) -> set[Any]:
    """
    Definieer het symmetrishe verschil van 2 verzamelingen in Python.
    """
    symmetrisch_verschil: set[Any] = A.symmetric_difference(B)

    return symmetrisch_verschil


def vraag3(n: int, a: int, b: int) -> set[int]:
    """
    Definieer de verzamelingen getallen groter dan 0 en kleiner dan $n$ die deelbaar zijn door $a$ maar niet door $b$.
    """
    verzameling: set[int] = {getal for getal in range(0, 100) if (getal > 0) and (getal < n) and ((getal % a == 0) and (getal % b != 0))}

    return verzameling


def jaccard(A: set[Any], B: set[Any]) -> float:
    """
    Geeft de Jaccard van twee verzamelingen terug.
    """
    A_en_B: set[Any] = A.intersection(B)
    A_of_B: set[Any] = A.union(B)
    
    jaccard: float = len(A_en_B) / len(A_of_B)

    return jaccard

```

```python
A = set([1,2,3])
B=set([4,5])
assert isinstance(product(A,B), set)
assert isinstance(symmetriesch_verschil(A,B), set)
assert isinstance(vraag3(1000,4,15), set)
assert isinstance(jaccard(A,B), float)
```

# Maak modellen

Schrijf de functie alle modellen, die gegeven een getal $n$ alle mogelijke modellen maakt voor $n$ propositieletters. 

Een model voor $n$ letters kan je makkelijk weergeven met een lijst van lengte $n$, gevuld met True/False of nullen en enen. De $i$-de positie geeft dan de waarheidswaarde van propositie $p_i$.

Je antwoord is dus een lijst van allemaal even lange lijsten.

**Hint** Het antwoord voor n=0 is dus geweoon `[[]]`.

```python
from itertools import product


def models(n: int) -> list[list[int]]:
    """
    Geeft het aantal mogelijke modellen voor n proposities letters terug.
    """
    if n == 0:
        return [[]]
    else:
        waarden: tuple[int, int] = [True, False]

        modellen: list[tuple[int]] = [list(rij) for rij in list(product(waarden, repeat=n))]
        
        return modellen
        

models(1), models(2), models(3)
```

2. Alle knopen wiens eccentricity gelijk is aan de diameter.
3.  Alle knopen wiens eccentricity gelijk is aan de radius
4. De knopen met degree 1.
5. De verzameling van alle edges die NIET in het netwerk zitten.
    * geen reflexieve edges
    * geen symmetrische

```python
v1: set[int] = set(knoop for knoop in G if G.degree(knoop) > 2)
v2: set[int] = set(knoop for knoop in G if nx.eccentricity(G, knoop) == nx.diameter(G))
v3: set[int] = set(knoop for knoop in G if nx.eccentricity(G, knoop) == nx.radius(G))
v4: set[int] = set(knoop for knoop in G if G.degree(knoop) == 1)
v5: set[int] = set()

for verbinding in G.edges():
    knoop1: int = verbinding[0]
    knoop2: int = verbinding[1]

    verbinding_omgekeerde_richting: tuple[int, int] = (knoop2, knoop1)

    if not ((knoop1 == knoop2) and (verbinding_omgekeerde_richting in G.edges())):
        v5.add(verbinding)

N=len(G)
v1, v2,v3,v4, len(v5), G.number_of_edges(), N*(N-1)//2,  len(v5)/(N*(N-1)//2), nx.density(G)
```

```python
from itertools import combinations


def hoogste_kans(G: nx.Graph) -> set[int]:
    """
    Geeft een paar knopen uit G waartussen geen lijn ligt en die de hoogste kans hebben om een i een lijn
    te krijgen volgens de triadicclosuretheorie. Als er meerdere zijn voldoet een willekeurige winnaar.
    """
    kanidaten: dict[tuple[int, int], int] = {}

    knoopcombinaties: set[int] = set(combinations(G.nodes(), 2))

    for knoop1, knoop2 in knoopcombinaties:
        gemeenschappelijke_buren: set[int] = set(nx.common_neighbors(G, knoop1, knoop2))
        kanidaten[(knoop1, knoop2)] = len(gemeenschappelijke_buren)
        
    beste_kanidaten: int = max(kanidaten, key=kanidaten.get)
    
    return beste_kanidaten



G: nx.Graph = nx.karate_club_graph()
nx.draw_kamada_kawai(G, with_labels=True)

hoogste_kans(G)
```

```python
# test hoeveel ie sluit
len(set(G.neighbors(2))& set(G.neighbors(33)))
### BEGIN HIDDEN TESTS
assert_equal(hoogste_kans(G), (2, 33))
### END HIDDEN TESTS
```

De neighborhood overlap (NO) van een lijn tussen twee knopen $A$ en $B$ is het aantal gedeelde buren van zowel A als B gedeeld door het totaal aantal buren van A en B min 2 (want we tellen A en B zelf niet mee).

$$Jaccard(A,B)= \frac{|A\cap B|}{|A\cup B|}$$

     1. $CC(A)=CC(B)=1$ en $NO((A,B))=0$.

##  Gatekeeper

Een knoop z is een *gatekeeper* voor knopen x en y in graaf G als z op elk kortste pad tussen x en y ligt. (Hier zijn x,y,z altijd verschillend).

En z is een gatekeeper in graaf G als z een gatekeeper is voor minstens twee knopen x en y.


Dus in een "spin in het web netwerk" met 1 middelpunt en allemaal ondeling niet verbonden satelitten is dat middelpunt een gatekeeper voor elk paar satelieten, en het is ook de enige gatekeper.

Definieer die functie en maak een paar mooie tests.

**HINT** Denk eens aan `nx.all_shortest_paths`.

**HINT** Dit is best een lastige. Je zou om te beginnen eerst een functie `gatekeep(G,z,x,y)` kunnen maken die `true` teruggeeft precies als `z` een gatkeeper is voor `x` en `y`. Dan kan je deze functie gebruiken in je antwoord.



```python


def is_gatekeeper(G: nx.Graph, mogelijke_gatekeeper: int, knoop1: int, knoop2: int) -> bool:
    """
    Geeft aan of een knoop een gatekeeper is in graaf G — of de knoop op elk pad kortste pad ligt.
    """
    kortste_paden: list[list[int]] = list(nx.all_shortest_paths(G, knoop1, knoop2))

    for kortste_pad in kortste_paden:
        if mogelijke_gatekeeper not in kortste_pad:
            return False

    return True

       
def gatekeepers(G: nx.Graph) -> set[int]:
    """
    Geeft een verzameling gatekeepers van een graaf G terug: knopen die tweemaal op elk kortste pad liggen voor een paar knopen x en y.
    """
    gatekeepers_G: set[int] = set()
    
    for mogelijke_gatekeeper in G:
        aantal_keer_gatekeeper: int = 0
        
        for knoop1 in G:
            for knoop2 in G:
                if len({knoop1, knoop2, mogelijke_gatekeeper}) == 3:  # alle drie uniek
                    if (is_gatekeeper(G, mogelijke_gatekeeper, knoop1, knoop2)):
                        aantal_keer_gatekeeper += 1
        
                        if aantal_keer_gatekeeper >= 2:
                            gatekeepers_G.add(mogelijke_gatekeeper)
                            break

    return gatekeepers_G


G= nx.karate_club_graph()    
gatekeepers(G)
```

```python
### BEGIN HIDDEN TESTS
assert_equal(gatekeepers(G), {0, 1, 2, 5, 6, 13, 23, 27, 31, 32, 33})
### END HIDDEN TESTS
```

# gatekeeper 2

* Nu willen we niet alleen de gatekeepers weten , maar ook voor hoeveel paren x,y ze gatekeeper zijn. Let op dat je paartjes maar 1 keer telt (dat kan bijv door te eisen dat voor een paartje (x,y) altijd `x<y` geldt.)
* Geef je antwoord als dict met de gatekeepers als sleutels en het aantal paren als waarde.

* Met wat geluk is het een kleine aanpassing van je eerdere code


```python
def gatekeepers(G: nx.Graph) -> set[int]:
    """
    Geeft een dictionary gatekeepers met hun frequentie als gatekeeper in een graaf G terug:
    knopen die tweemaal op elk kortste pad liggen voor een paar knopen x en y.
    """
    gatekeepers_met_frequentie: dict[int, int] = {}
    
    for mogelijke_gatekeeper in G:
        aantal_keer_gatekeeper: int = 0
        
        for knoop1 in G:
            for knoop2 in G:
                if len({knoop1, knoop2, mogelijke_gatekeeper}) == 3:  # alle drie uniek
                    if (is_gatekeeper(G, mogelijke_gatekeeper, knoop1, knoop2)):
                        if not mogelijke_gatekeeper in gatekeepers_met_frequentie.keys():
                            gatekeepers_met_frequentie[mogelijke_gatekeeper] = 1
                        else:
                            gatekeepers_met_frequentie[mogelijke_gatekeeper] += 1

    gatekeepers_met_frequentie: dict[int, int] = {
        gatekeeper: frequentie // 2 for gatekeeper, frequentie
        in gatekeepers_met_frequentie.items() if frequentie >= 2
    } 
    
    return gatekeepers_met_frequentie
    
gatekeepers(G)
```
