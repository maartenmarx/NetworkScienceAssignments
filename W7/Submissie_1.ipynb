## Antwoord
1. De *closeness centrality* van een knoop laat zien hoe dichtbij een knoop bij alle andere knopen in het netwerk ligt.
2. Als je een hoge cc hebt dan lig je dichtbij alle andere knopen in het netwerk, als je een lage cc hebt lig je ver weg van alle andere knopen in het netwerk en zit je waarschijnlijk aan de rand van een netwerk.
