```python
G=nx.gn_graph(20,seed=9) # kan ook met 10 ipv 20
for n in G:
    if n%2==0:
        G.nodes[n]['P']=True
    else:
        G.nodes[n]['P']=False
for n in G:
    if n%3==0:
        G.nodes[n]['Q']=True
    else:
        G.nodes[n]['Q']=False        
print(G.nodes(data=True))        
nx.draw_networkx(G)
```

```python
G=nx.karate_club_graph() # kan ook met 10 ipv 20

for n in G:
    if G.nodes[n]['club']=='Mr. Hi':
        G.nodes[n]['P']=True
        G.nodes[n]['Q']=False  
    else:
        G.nodes[n]['P']=False
        G.nodes[n]['Q']=True
    del G.nodes[n]['club']    
print(G.nodes(data=True))        
nx.draw_networkx(G)
```

```python
# even wat afkortingen

def P(x):
    return G.nodes[x]['P']
def Q(x):
    return G.nodes[x]['Q']
def R(x,y):
    return G.has_edge(x,y)
```

```python
## Voorbeeld

# R(x,y) betekent x houdt van y  
# Bij de karate club is de relatie ongericht en kan je beter van x is vriend met y spreken.
# Er is een knoop waar zowel een P knoop als een Q knoop van houdt.
# $\exists x\exists y\exists z(R(y,x)\wedge R(z,x) \wedge P(y)\wedge Q(z))$

print(any(R(y,x) and  R(z,x) and  P(y) and  Q(z)  for x in G for y in G for z in G))


def LoveByAPandAQ(x):
    '''return True iff zowel een P knoop als een Q knoop van x  houdt'''
    return any(R(y,x) and  R(z,x) and  P(y) and  Q(z)   for y in G for z in G)


print({knoop for knoop in G if LoveByAPandAQ(knoop)}) # geef alle x met die eigenschap


# en welke zijn het nou, en waarom      
allemaal= [ (x,y,z)   for x in G for y in G for z in G 
    if R(y,x) and  R(z,x) and  P(y) and  Q(z)
]  

len(allemaal), allemaal[:5]
```
