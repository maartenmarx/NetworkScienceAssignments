## Vraag 5

* Maak een functie `find_SCC` die gegeven een gerichte graaf een set teruggeeft met daarin alle nodes die behoren tot de grootste SCC (strongly connected component).

* Maak een functie `find_IN` die gegeven een gerichte graaf een set teruggeeft met daarin alle nodes die behoren tot de IN (dat zijn alle knopen met een pad naar de SCC, maar vanuit de SCC kan je niet naar die knopen komen).

* Maak een functie `find_OUT` die gegeven een gerichte graaf een set teruggeeft met daarin alle nodes die behoren tot de OUT (ja, dit snap je nu zelf wel hoop ik).

```python
def find_SCC(G: nx.DiGraph) -> set[int]:
    """
    Ontvangt een gerichte graaf, en geeft een verzameling terug waarin alle knopen zitten
    die behoren tot het sterkstverbonden component ('SCC'; 'strongly connected component').
    """
    sterkstverbonden_component: int = set(max(nx.strongly_connected_components(G), key=len))

    return sterkstverbonden_component


def find_OUT(G: nx.DiGraph) -> set[int]:
    """
    Ontvangt een gerichte graaf, en geeft een verzameling knopen terug die behoren tot de uit-graad van de scc: 
    alle knopen zonder pad naar de scc, maar waar de scc wel een verbinding mee heeft.
    """
    ssc_knopen: set[int] = find_SCC(G)
    knopen_alleen_bereikbaar_voor_scc: set[int] = {knoop for knoop in G if nx.has_path(G, knoop, list(ssc_knopen)[0]) and knoop not in ssc_knopen}

    return knopen_alleen_bereikbaar_voor_scc
    

def find_IN(G: nx.DiGraph) -> set[int]:
    """
    Ontvangt een gerichte graaf, en geeft een verzameling knopen terug die behoren tot de in-graad van de scc: 
    alle knopen met een pad naar de scc, maar die vanuit de scc niet bereikt kunnen worden.
    """
    ssc_knopen: set[int] = find_SCC(G)
    knopen_onbereikbaar_voor_scc: set[int] = {knoop for knoop in G if nx.has_path(G, list(ssc_knopen)[0], knoop) and knoop not in ssc_knopen}

    return knopen_onbereikbaar_voor_scc
    

find_SCC(EK), find_IN(EK), find_OUT(EK)
```

```python
assert_equal(type(find_SCC(EK)), set)
assert_equal(type(find_OUT(EK)), set)
assert_equal(type(find_IN(EK)), set)
assert find_SCC(EK).issubset(EK.nodes())
assert find_IN(EK).issubset(EK.nodes())
assert find_OUT(EK).issubset(EK.nodes())
```

```python
# resultaat graaf tekenen
def draw_web(G):
    pos = nx.spring_layout(G)
    nx.draw(G, pos, with_labels=True)
    color = sns.color_palette("hls", 7)
    #tubes = find_tubes(G)
    legenda = [
        'SCC', 'IN', 'OUT', ] #'Tendrils (IN)', 'Tendrils (OUT)', 'Tubes','Disconnected' ]
    listofnodes = [
        find_SCC(G),
        find_IN(G),
        find_OUT(G),
        #find_tendrils_IN(G) - tubes,
        #find_tendrils_OUT(G) - tubes, tubes,
        #find_disconnected(G)
    ]
    for n, (label, nodes, c) in enumerate(zip(legenda, listofnodes, color)):
        nx.draw_networkx_nodes(
            G, pos, nodelist=nodes, node_color=[c], label=label)
    nx.draw_networkx_edges(G, pos)
    plt.legend()
    plt.axis('off')
    plt.show()


draw_web(EK)
```

## Vraag 3.1
Maak een functie `clusteringcoefficient(G)` die een graaf accepteert als input, en voor alle knopen de clusteringcoëfficiënt ([Sectie 3.1 E&amp;K](http://www.cs.cornell.edu/home/kleinber/networks-book/networks-book-ch03.pdf)) berekent.
                
<strong>Bereken dit zelf, gebruik geen NetworkX functie die dit al doet zoals `clustering` en `average_clustering`. </strong> Je mag natuurlijk wel NetworkX handigheidjes als `G.neighbors(v)` gebruiken.

Als output krijg je een dict  met de knopen van `G` als sleutels en de clustering coeffiecient van die knoop als bijbehorende waarde.   Geef waarde 0 als een knoop degree 1 heeft.


De verwachte output voor deze graaf is 1 voor knoop 7 en 2/3 voor knoop 3 en 0 voor knoop 11. etc. 
  

```python
def tel_aantal_unieke_driehoeken(G: nx.Graph, buren_van_knoop: set[int]) -> int:
    """
    Ontvangt een graaf en een verzameling buren, en geeft het aantal driehoeken van deze buren terug.
    """
    aantal_driehoeken: int = 0
    
    for knoop1 in buren_van_knoop:
        for knoop2 in buren_van_knoop:
            if knoop1 < knoop2:  # voorkomt dubbele driehoeken
                if G.has_edge(knoop1, knoop2) and G.has_edge(knoop2, knoop1):
                    aantal_driehoeken += 1
    
    return aantal_driehoeken


def clusteringcoefficient(G: nx.Graph) -> dict[int, float]:
    """
    Ontvangt een graaf, en geeft de lokale clustervormingscoëfficiënt van iedere knoop in deze graaf terug.
    """
    knopen_met_clustervormingscoefficient: dict[int, float] = {}
    
    for knoop in G.nodes():
        buren_van_knoop: set[int] = set(G.neighbors(knoop))
        graden_knoop: int = G.degree(knoop)

        if graden_knoop < 2:
            knopen_met_clustervormingscoefficient[knoop] = 0.0
        else:
            aantal_driehoeken_onder_buren: int = tel_aantal_unieke_driehoeken(G, buren_van_knoop)
            clustervormingscoefficient: float = aantal_driehoeken_onder_buren / (graden_knoop * (graden_knoop - 1) / 2)
            knopen_met_clustervormingscoefficient[knoop] = clustervormingscoefficient

    return knopen_met_clustervormingscoefficient


# even een mooi plotje van de distributie van de CCs maken
A= pd.Series( clusteringcoefficient(G) )
sns.displot(A)

clusteringcoefficient(G)
    
```

```python
assert isinstance(clusteringcoefficient(G), dict)
for n in clusteringcoefficient(G):
    assert n in G
    assert isinstance(clusteringcoefficient(G)[n],float)
```

## Vraag 3.1.2

Maak de functie `cc_nicer(G, v)` nu nog een keer maar gebruik nu alleen maar  `G.subgraph` en `nx.density` van networkx.
Check dat hij echt hetzelfde doet als je eerdere functie en als networkx clustering.

```python
def cc_nicer(G: nx.Graph, knoop: int) -> float:
    """
    Ontvangt een graaf en een knoop, en berekent de clustervormingscoëfficiënt van de knoop in deze graaf.
    """
    buren_van_knoop: list[int] = [buur for buur in G if G.has_edge(buur, knoop)]  
    subgraaf: G.subgraph = G.subgraph(buren_van_knoop)
    
    clustervormingscoefficient: float = nx.density(subgraaf)
    
    return clustervormingscoefficient


# check dat de twee functies hetzelfde resultaat geven voor elke knoop op de karateclub
clusteringcoefficient(G)== {n:cc_nicer(G,n) for n in G}
```

2. Maak de functie  `is_complete(G)`  die `True` teruggeeft precies dan als het ongerichte netwerk  `G` een volledig netwerk is en `False` als dat niet het geval is.  Gebruik **alleen** de networkx functies `nx.number_of_nodes(G)` en `G.number_of_edges()`.

```python
def is_complete(G: nx.Graph) -> bool:
    """
    Ontvangt een ongerichte graaf, en geeft terug of deze compleet is: n = (n * (n - 1)) / 2.
    """
    aantal_knopen_graaf: int = nx.number_of_nodes(G)
    aantal_verbindingen_graaf: int = nx.number_of_edges(G)

    aantal_knopen_complete_graaf: int = int((aantal_knopen_graaf * (aantal_knopen_graaf - 1)) / 2)
    
    return aantal_verbindingen_graaf == aantal_knopen_complete_graaf


is_complete(G)
```
