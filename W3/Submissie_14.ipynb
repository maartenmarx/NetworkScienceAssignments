```python
def find_SCC(G):

    #elke node in een SCC heeft het zelfde aantal nodes waar het heen en weer kan, wanneer er een dictionary wordt gamaakt, die voor elke
    #node kijkt naar hoeveel andere nodes deze node kan + of die andere nodes deze node kunnen bereiken, zal er een cluster vormen, zullen
    #de nodes met de grootste waarde in deze dictionary behoren tot de cluster.
    connected_level = {node1: sum([nx.has_path(EK, node1, node2) and nx.has_path(EK, node2, node1) for node2 in EK]) for node1 in EK}
    highest_value = max(connected_level.values())
    SCC = {node for node in connected_level if connected_level[node] == highest_value}
    if (len(SCC) / highest_value) <= 1:
        return SCC
    elif (len(SCC) / highest_value) > 1:
        more_SCC = {set(bfs_tree(EK,n1,n2)) for n1 in SCC
                        for n2 in SCC}
        return "Er zijn meerdere SCC's", more_SCC


def find_OUT(G):
    out_of_cluster = {node for nodes_SCC in find_SCC(EK) for node in EK 
                      if nx.has_path(EK, nodes_SCC, node) and not nx.has_path(EK, node, nodes_SCC)}

    return out_of_cluster

def find_IN(G):
    in_of_cluster = {node for nodes_SCC in find_SCC(EK) for node in EK 
                      if nx.has_path(EK, node, nodes_SCC) and not nx.has_path(EK, nodes_SCC, node)}

    return in_of_cluster
    
# even testen op figuur 13.8 uit het boek
find_SCC(EK), find_IN(EK), find_OUT(EK)
```

## Vraag 3.1
Maak een functie `clusteringcoefficient(G)` die een graaf accepteert als input, en voor alle knopen de clusteringcoëfficiënt ([Sectie 3.1 E&amp;K](http://www.cs.cornell.edu/home/kleinber/networks-book/networks-book-ch03.pdf)) berekent.
                
<strong>Bereken dit zelf, gebruik geen NetworkX functie die dit al doet zoals `clustering` en `average_clustering`. </strong> Je mag natuurlijk wel NetworkX handigheidjes als `G.neighbors(v)` gebruiken.

Als output krijg je een dict  met de knopen van `G` als sleutels en de clustering coeffiecient van die knoop als bijbehorende waarde.   Geef waarde 0 als een knoop degree 1 heeft.


De verwachte output voor deze graaf is 1 voor knoop 7 en 2/3 voor knoop 3 en 0 voor knoop 11. etc. 
  

```python

def clusteringcoefficient(G):

    clustering_coefficient = {node: (sum(len(set(G.neighbors(buur)) & set(G.neighbors(node))) for buur in G.neighbors(node)) /
            (len(buren) * (len(buren) - 1)) if len(buren) > 1 else 0.0)
        for node in G
        for buren in [list(G.neighbors(node))]}

            
    return clustering_coefficient


#even een mooi plotje van de distributie van de CCs maken
A= pd.Series( clusteringcoefficient(G) )
sns.displot(A)

clusteringcoefficient(G)
```

```python
assert isinstance(clusteringcoefficient(G), dict)
for n in clusteringcoefficient(G):
    assert n in G
    assert isinstance(clusteringcoefficient(G)[n],float)
```

## Vraag 3.1.1

Maak de functie `cc_nice(G, v)` nu nog een keer maar gebruik nu alleen maar `G.has_edge()` van networkx.
Chack dat hij echt hetzelfde doet als je eerdere functie en als networkx clustering.

```python
def cc_nice(G, v):
    has_edges_to = [node for node in G if G.has_edge(v, node) == True]
    triangle_succes = sum([G.has_edge(n1, n2) for n1 in has_edges_to for n2 in has_edges_to])
    return triangle_succes / (len(has_edges_to) * (len(has_edges_to) - 1)) if len(has_edges_to) > 1 else 0.0

# check dat de twee functies hetzelfde resultaat geven voor elke knoop op de karateclub
clusteringcoefficient(G)== {n:cc_nice(G,n) for n in G}
```

2. Maak de functie  `is_complete(G)`  die `True` teruggeeft precies dan als het ongerichte netwerk  `G` een volledig netwerk is en `False` als dat niet het geval is.  Gebruik **alleen** de networkx functies `nx.number_of_nodes(G)` en `G.number_of_edges()`.

```python
def is_complete(G):
    N_n = nx.number_of_nodes(G)
    N_e = G.number_of_edges()
    return N_e == N_n * (N_n - 1) / 2
is_complete(G)
```
